```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_excel(r"Titanic_dataset.xlsx")

print(df)
```

```python

sns.countplot(x='survived', data=df)
plt.title('Survival Distribution')
plt.show()
```

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival Distribution by Gender')
plt.show()
```

```python
# Visualize the survival count based on passenger class
plt.figure(figsize=(8, 5))
sns.countplot(x='pclass', hue='survived', data=df, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Visualize the age distribution of passengers
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), kde=False, bins=30, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the distribution of fare paid
plt.figure(figsize=(10, 6))
sns.histplot(df['fare'], kde=False, bins=30, color='green')
plt.title('Distribution of Fare Paid by Passengers')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# Pair plot to visualize relationships between numerical features
sns.pairplot(df, hue='survived', palette='Set1')
plt.suptitle('Pair Plot of Numerical Features by Survival', y=1.02)
plt.show()
```

```python
# Violin plot to visualize the distribution of age by class and survival
plt.figure(figsize=(12, 8))
sns.violinplot(x='pclass', y='age', hue='survived', data=df, split=True, palette='Set2')
plt.title('Distribution of Age by Class and Survival')
plt.show()
```

```python
# Catplot to visualize survival rates by class and sex
plt.figure(figsize=(10, 6))
sns.catplot(x='pclass', y='survived', hue='sex', data=df, kind='bar', palette='Set3')
plt.title('Survival Rates by Class and Sex')
plt.show()
```

```python
# Swarm plot to visualize the distribution of ages by class and survival
plt.figure(figsize=(12, 8))
sns.swarmplot(x='pclass', y='age', hue='survived', data=df, palette='Set2')
plt.title('Swarm Plot of Age Distribution by Class and Survival')
plt.show()
```

```python
# Heatmap to visualize missing data
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data: Yellow indicates missing values')
plt.show()
```

```python
# Pie chart to show the distribution of passengers by class
class_counts = df['pclass'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Passengers by Class')
plt.show()
```
